```python
from dask_ml import preprocessing

ddf = ddf.categorize(columns=['D_63', 'D_64'])
encoder = preprocessing.DummyEncoder(columns=['D_63', 'D_64'])  # OneHot
ddf = encoder.fit_transform(ddf)  # it's ok for group by mean below 
```

```python
ddf_grouped = ddf.groupby(by='customer_ID').mean()  # it's bad with nan
```

```python
# add max and min time
ddf_grouped = ddf_grouped.join(ddf.groupby(by='customer_ID')['S_2'].max(), how='left'
                  ).rename(columns={'S_2': 'S_2_max'}).join(ddf.groupby(by='customer_ID')['S_2'].min()
                                                           ).rename(columns={'S_2': 'S_2_min'})

ddf_grouped = ddf_grouped.join(ddf.groupby(by='customer_ID')['S_2'].count(), how='left').rename(columns={'S_2': 'count'})
```

```python
# delta max and min data
def my_delta_time(x):
    try:
        return int((datetime.datetime.strptime(x['S_2_max'], "%Y-%m-%d") - datetime.datetime.strptime(x['S_2_min'], "%Y-%m-%d")).days)
    except ValueError:
        return -1

ddf_grouped['S_2_delta'] = ddf_grouped.apply(my_delta_time, axis=1)
ddf_grouped = ddf_grouped.drop(['S_2_max', 'S_2_min'], axis=1)
```

# Model

```python
import lightgbm as lgbm
import dask_ml.model_selection
import os
from sklearn.metrics import roc_auc_score

# join with targets
ddf_grouped = ddf_grouped.join(ddt.set_index('customer_ID'), how='inner')
```

```python
# split on train, val and test
# It's necessary to split after all groups
X_train, X_test, y_train, y_test = dask_ml.model_selection.train_test_split(ddf_grouped.drop(['target'], axis=1),
                                                                            ddf_grouped['target'],
                                                                            test_size=0.2,
                                                                            shuffle=True,
                                                                            random_state=42)  # there is no stratify

X_train, X_val, y_train, y_val = dask_ml.model_selection.train_test_split(X_train,
                                                                          y_train,
                                                                          test_size=0.25,
                                                                          shuffle=True,
                                                                          random_state=42)  # there is no stratify
```
